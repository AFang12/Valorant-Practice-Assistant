```python
def read_key_log(key_log_path):
    formatted_list = []

    # 打开文件并读取内容
    with open(key_log_path, 'r') as file:
        for line in file:
            # 去除行末的换行符
            line = line.strip()
            # 通过', '分割字符串，得到键和值的部分
            parts = line.split(', ')
            # 获取键值对
            key = parts[0].split(': ')[1]
            timestamp = float(parts[1].split(': ')[1])
            # 将键值对作为列表添加到结果列表中
            formatted_list.append([key, timestamp])
    return formatted_list
```
